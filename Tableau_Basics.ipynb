Q1. Explain what Tableau is and why it is preferred over traditional spreadsheet tools for data analysis.

Also mention any two industries where Tableau is commonly used and explain one use case.

-> Tableau is a business intelligence tool for creating interactive visualizations from diverse data sources. It is preferred over spreadsheets like Excel for handling large datasets without lag, offering drag-and-drop interactivity, and enabling real-time dashboards for deeper insights. Two industries: Retail (e.g., analyzing sales trends by region to optimize stock) and Healthcare (tracking patient metrics for operational efficiency).
